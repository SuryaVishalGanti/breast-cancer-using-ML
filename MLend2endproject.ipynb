```python
## Independent & Dependent Feature,

X = data.iloc[:,::-1]
y = data.iloc[:,-1]
```

```python
y
```

```python
# Trainig the Model,

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
## Model Training
## Standardizing the dataset
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
X_train = scale.fit_transform(X_train)
```

```python
X_test = scale.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Training our Model

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## Printing the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## On which Parameters the model has been trained
regression.get_params()
```

```python
## Prediction with test data
reg_pred = regression.predict(X_test)
```

```python
## we can definaly compare with y_test
## Plotting a scatter plot for Prediction
plt.scatter(reg_pred, y_test)
```

```python
## Prediction with residual
residual = y_test-reg_pred
residual
```

```python
## Plotting this residuals

sns.displot(residual, kind='kde')
```

```python
## Assuming another scatter plot and residuals
## Uniform distribution
plt.scatter(reg_pred, residual)
```

```python
## Performance Matrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

## Printing the mean_absolute error
print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

## R square and adjusted square

Formula

#### **R^2 = 1-SSR/SST**

R^2 = coefficient of determination, SSR = sum of squares of residuals, SST = total sum of squares

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

### **Adjusted R^2 = 1 - [(1 - R2) * (n/1)/(n-k-1)]**

```python
## Displaying the adjusted R- squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
